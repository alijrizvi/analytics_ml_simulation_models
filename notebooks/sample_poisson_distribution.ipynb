```python
from scipy.stats import poisson
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating a Random Set of Goals Scored per Game

x_rvs = pd.Series(poisson.rvs(mu = 1.4, size = 1000)) # "mu" = average value, "size" = rows/matches (data points)
```

```python
# Organizing the Values for Subsequent Plotting

data = x_rvs.value_counts().sort_index().to_dict()
data
```

```python
fig, ax = plt.subplots(figsize = (14, 6))

ax.bar(range(len(data)), list(data.values()), align = 'center')
plt.title('Probability of Goals Scored per Game')
plt.xlabel('Goals')
plt.ylabel('Count of Occurrence')
plt.xticks(range(len(data)), list(data.keys()))
plt.show()
```

```python
poisson.pmf(2, 1.4) # Wanting to Calculate the Probability of a Team Scoring 2 Goals in a Match, Despite knowing that, on Average, 1.4 Goals are Scored

# Answer: 24.2%
```

```python
df = pd.DataFrame(data, index = [0])
df = df.T.rename(columns = {0: 'Amount of Times These Many Goals Were Scored'})
df['Serial Number'] = np.random.randint(1000, 9999, 8)

df
```

```python
# Was there any Relationship Between Serial Number of Goal Count & The Occurrence of that Goal Count?

import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize = (15, 5))

# jointplot is a figure-level function and does not accept the ax parameter
sns.lineplot(x = 'Serial Number', y = 'Amount of Times These Many Goals Were Scored', data = df, ax = axes[0])

# relplot is a figure-level function, using scatterplot which is axes-level
sns.barplot(x = 'Serial Number', y = 'Amount of Times These Many Goals Were Scored', data = df, ax = axes[1])

## sns.jointplot(x = , y = , data = )
## sns.relplot(x = , y = , data = )

plt.show()
```
